```python
import scipy
from scipy import stats
import seaborn as sns
%matplotlib inline
```

### Using your covariance matrix, Generate from multivariate Gaussian


```python
mvnorm = stats.multivariate_normal(mean=[0, 0], cov=[[1., 0.7], 
                                                     [0.7, 1.]])
# Generate random samples from multivariate normal with correlation .5
x = mvnorm.rvs(10000)
```

```python
h = sns.jointplot(x[:, 0], x[:, 1], kind='kde', stat_func=None);
h.set_axis_labels('X1', 'X2', fontsize=16);
```

### Inverse transform 

U = norm.cdf(norm)

D = D.inverse_cdf(U)

https://en.wikipedia.org/wiki/Inverse_transform_sampling

```python
norm = stats.norm()
x_unif = norm.cdf(x)

h = sns.jointplot(x_unif[:, 0], x_unif[:, 1], kind='hex', stat_func=None)
```

 ### transform each marginal to desired distribution of that column
    
    Sklar theorem 


<img src="marginal_conditional.png">,

```python
m1 = stats.gumbel_l()  # Age column
m2 = stats.beta(a=10, b=2)  # Speed column
```

### this is your synthetic data with desired distribution and covariance


```python
x1_trans = m1.ppf(x_unif[:, 0])
x2_trans = m2.ppf(x_unif[:, 1])

```

```python
import pandas as pd

df = pd.DataFrame({'x1' : x1_trans, 'x2' : x2_trans})
```

### matches original covariance above

```python
print(df.corr().values)
```
